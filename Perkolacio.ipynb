```python
import random
import matplotlib.pyplot as plt
import numpy as np
from numba import jit

def create_grid(n):
    grid = [[False for i in range(n)] for j in range(n)]
    return grid

def site_percolation(grid, p):
    n = len(grid)
    for i in range(n):
        for j in range(n):
            if random.random() < p:
                grid[i][j] = True

class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.size = [1] * n
        self.max_cluster_size = 1
    
    def find(self, p):
        while p != self.parent[p]:
            self.parent[p] = self.parent[self.parent[p]]
            p = self.parent[p]
        return p
    
    def union(self, p, q):
        root_p = self.find(p)
        root_q = self.find(q)
        if root_p == root_q:
            return
        if self.size[root_p] < self.size[root_q]:
            self.parent[root_p] = root_q
            self.size[root_q] += self.size[root_p]
            self.update_max_cluster_size(root_q)
        else:
            self.parent[root_q] = root_p
            self.size[root_p] += self.size[root_q]
            self.update_max_cluster_size(root_p)

    
    def get_cluster_size(self, p):
        root_p = self.find(p)
        return self.size[root_p]
    
    def update_max_cluster_size(self, root):
        if self.size[root] > self.max_cluster_size:
            self.max_cluster_size = self.size[root]
    
    
    def find_percolating_cluster(self):
        n = int(np.sqrt(len(self.parent)))
        top_row = set(self.find(i) for i in range(n*n) if i < n)
        bottom_row = set(self.find(i) for i in range(n*n) if i >= n * (n - 1))
        percolating_cluster = top_row & bottom_row
        return percolating_cluster

def simulate_site_percolation(n, p_values, perkolacio=False):
    results = []
    for p in p_values:
        grid = create_grid(n)
        site_percolation(grid, p)
        uf = UnionFind(n * n)

        for i in range(n):
            for j in range(n):
                if grid[i][j]:
                    current_site = i * n + j
                    if i > 0 and grid[i - 1][j]:
                        uf.union(current_site, (i - 1) * n + j)
                    if i < n - 1 and grid[i + 1][j]:
                        uf.union(current_site, (i + 1) * n + j)
                    if j > 0 and grid[i][j - 1]:
                        uf.union(current_site, i * n + j - 1)
                    if j < n - 1 and grid[i][j + 1]:
                        uf.union(current_site, i * n + j + 1)
        #print(uf.size,uf.parent,uf.max_cluster_size)
        cluster_max_sizes = uf.max_cluster_size
        if perkolacio:
          percolating_clusters = uf.find_percolating_cluster()
          percolating_cluster_sizes = [uf.get_cluster_size(site) for site in percolating_clusters]
        else:
          percolating_clusters, percolating_cluster_sizes= None, None

        results.append((p, percolating_clusters, percolating_cluster_sizes, cluster_max_sizes))
    
    return results
```

```python
def proba(n = 100, num_experiments = 100, k = 10, kozelito=False):
  # Probabilities to test
  if kozelito:
    p_values = [0.55+i/(10*k) for i in list(range(k))]
  else:
    p_values = [i/k for i in list(range(k))]

  average_sizes = []

  for _ in range(num_experiments):
    results = simulate_site_percolation(n, p_values)
    
    # Accumulate cluster sizes for each p
    if not average_sizes:
        average_sizes = [cluster_sizes for _, _, _, cluster_sizes in results]
    else:
        average_sizes = [avg + cluster_sizes for avg, (_, _, _, cluster_sizes) in zip(average_sizes, results)]

  # Calculate average cluster sizes over experiments
  average_sizes = [np.mean(avg) / num_experiments for avg in average_sizes]
  return p_values, average_sizes
```
